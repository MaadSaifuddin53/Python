```python
def gcd(num1,num2):
    gcd=1
    i=1
    if num1 == 0 or num2 == 0:
        print(gcd)
    rem=num1%num2
    print('Iteraton # ',i,'------>',num1,'%',num2,'==',rem)
    str1=' Iteraton # '+str(i)+' ------> '+str(num1)+' % '+str(num2)+' == '+str(rem)+'\n'
    with open('file','a') as f:
        f.write(str1)
        f.write('\n')
        while rem != 0:
            i=i+1
            num1=num2
            num2=rem
            rem=num1%num2
            str1='Iteraton # '+str(i)+' ------> '+str(num1)+' % '+str(num2)+' == '+str(rem)+'\n'
            f.write(str1)
            f.write('\n')
        gcd=num2
        print('Number of basic operation perform',i)
        print(' GCD = ',gcd)
        str2=' Number of basic operation perform '+str(i)+'\n'
        str3=' GCD = '+str(gcd)+'\n'
        f.write(str2)
        f.write(str3)
        f.write('\n')
        f.close()
        return i
```

```python
       
count=0
sum1=0
for i in range(100,1000,10):
    for j in range(990,500,-20):
            count=count+1
            a=gcd(i,j)
            sum1=sum1+a       
```
